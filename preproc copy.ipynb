```python
import os
import pandas as pd
import warnings
import tqdm
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
to_delete=[]
for root, dirs, files in os.walk('./UNGDC_1970-2020/'):
    for file in files:
        if file.startswith('.'):
            to_delete.append(os.path.join(root, file))
            continue

for loc in to_delete:
    os.remove(loc)
```

```python
file_to_use=[]
for root, dirs, files in os.walk('./UNGDC_1970-2020/'):
    for file in files:
        if not file.startswith('.'):
            file_to_use.append(os.path.join(root, file))
```

```python
num_lines={}
for file in file_to_use:
    with open(file, 'r+', encoding='utf8') as infile:
        num_lines[file]=len(infile.readlines())
```

```python
total_num_lines = sum([item[1] for item in num_lines.items()])
```

```python
sorted_num_lines=sorted(num_lines.items(), key=lambda k: k[1])
```

```python
def get_lines(file_loc):
    count=0
    for line in open(file_loc, 'r+', encoding="utf8"):
        yield count, line
        count+=1
```

```python
file_count=0
sent_count=0
dict_list=[]
for file in file_to_use:
    temp_list=[]
    for count, line in get_lines(file):
        print(f'Working on \'{file}\'; \nTotal files done: {file_count}/{len(file_to_use)}; \nSentences done: {sent_count}/{total_num_lines}', end= '\r\r\r')
        temp_list.append({
            'doc_id':file+str(count+1), 
            'text':line, 
            'country':file.split('_')[0], 
            'year':file.split('_')[2].split('.')[0]
            })
        sent_count+=1
    file_count+=1
    dict_list.append(temp_list)
# pd.DataFrame(columns=['doc_id','text','country','year','coding'])
```

```python
count=0
sent_count=0
df=pd.DataFrame(columns=['doc_id','text','country','year','coding'])
for loc in file_to_use:
    with open(loc, 'r+', encoding="utf8") as infile:
        lines=infile.readlines()
        for idx, line in enumerate(lines):
            print(f'Working on \'{file}\'; \nTotal files done: {count}/{len(file_to_use)}; \nSentences done: {sent_count}/{total_num_lines}', end= '\r')
            df=df.append({'doc_id':file+str(idx+1), 'text':line, 'country':file.split('_')[0], 'year':file.split('_')[2].split('.')[0]}, ignore_index=True)
            sent_count+=1
        count+=1
```

```python
df.to_csv('data.csv')
```
